# Aceleração de convergência

## Questão 1: Implemente o [método de Aitken](https://en.wikipedia.org/wiki/Aitken%27s_delta-squared_process#Definition)

```python
def aitken(xs):
    """Sequência acelerada a partir de  x_n,  após aplicar a transformação de Aitken."""
    xs = np.array(xs)
    ### BEGIN SOLUTION
    dxs  = xs[1:] - xs[:-1]
    ddxs = dxs[1:] - dxs[:-1]
    ys = xs[:-2] - dxs[:-1]**2/ddxs
    return ys
    ### END SOLUTION
```

A sequência de Leibniz converge para $\pi/4$, mas devagar:

```python
ts = np.array([(-1)**i/(2*i + 1) for i in range(20)])
approx_pi = 4*np.cumsum(ts)
plt.plot(approx_pi - np.pi)
plt.title("Erro de aproximação")
plt.grid()
```

Vamos acelerar!

```python
acc_pi = aitken(approx_pi)

assert len(acc_pi) == 18
assert 1e-5 < np.abs(acc_pi[-1] - np.pi) < 1e-4
```

Como ficou o erro?
Faça um gráfico.

```python
### BEGIN SOLUTION
plt.plot(acc_pi - np.pi)
plt.title("Erro de aproximação após uma iteração")
plt.grid()
### END SOLUTION

ax = plt.gca()
plt.show()
```

## Questão 2: Muitas iterações

Repita o procedimento até não ser mais possível, e faça um gráfico com todas as sequências produzidas,
desde `approx_pi` até a última.

```python
### BEGIN SOLUTION
def aitken_plot(seq, ans, maxiter=100, ns=None, style="-"):
    cur = seq
    if ns is None:
        ns = np.arange(1,len(seq)+1)
    for i in range(maxiter+1):
        plt.semilogy(ns[2*i:], np.abs(cur - ans), style, label=str(i))
        if len(cur) <= 2: break
        cur = aitken(cur)
    plt.legend(title="# Acelerações", loc="center left", bbox_to_anchor=(1.0, 0.5))
    plt.grid()

aitken_plot(approx_pi, np.pi)

plt.title("Erro de aproximação")
plt.xlabel("Número de termos utilizados")
### END SOLUTION

ax = plt.gca()
plt.show()
```

```python
assert ax.title.get_text() != ""
assert len(ax.lines) == 10
assert len(ax.get_legend().get_texts()) == 10
```

```python
for l in ax.get_lines():
    xdata = l.get_xdata()
    assert max(xdata) == 20
```

## Questão 3: Acelerando os métodos clássicos

Sabemos que, ao aumentar o número de subdivisões, os métodos de integração se aproximam do valor exato da integral.
Aplique a aceleração de Aitken aos métodos:
- do ponto médio,
- de Simpson,

para aproximar a integral $$\int_0^2 \sin(x)\, dx ,$$
aumentando o número de subdivisões.

```python
ans = 1 - np.cos(2)
```

```python
def simpson(f,a,b, n):
    """ Simpson's quadrature: integrates  f  over the interval  [a,b]  with  n  subdivisions."""
    ### BEGIN SOLUTION
    xs, h = np.linspace(a,b, num=n, endpoint=False, retstep=True)
    mids = f(xs+h/2)
    ints = f(xs[1:])

    return (4*np.sum(mids) + (2*np.sum(ints) + (f(a) + f(b))))*h/6
    ### END SOLUTION

def midpoint(f,a,b, n):
    """ Midpoint quadrature: integrates  f  over the interval  [a,b]  with  n  subdivisions."""
    ### BEGIN SOLUTION
    xs, h = np.linspace(a,b, num=n, endpoint=False, retstep=True)
    mids = f(xs+h/2)

    return np.sum(mids)*h
    ### END SOLUTION
```

## Questão 3.1: Qual subsequência acelerar?

Calcule `I_midp` e `I_simp` para valores de $n$ de 1 a 1024.

```python
I_midp = [midpoint(np.sin, 0, 2, n) for n in range(1,1024)]
I_simp = [simpson(np.sin, 0, 2, n) for n in range(1,1024)]
```

```python
assert len(I_midp) == 1024-1
assert len(I_simp) == 1024-1
```

```python
assert not np.isclose(I_midp[100], ans)
assert not np.isclose(I_simp[3], ans)
assert np.isclose(I_midp[200], ans)
assert np.isclose(I_simp[4], ans)
```

```python
acc_midp = aitken(I_midp)
assert np.isclose(acc_midp[98], ans)
acc_simp = aitken(I_simp)
assert np.isclose(acc_simp[2], ans)
```

Acelere ambas as sequências, 2 vezes, e faça um gráfico para o erro.

## Questão 4.3: Domínio infinito

Agora, desejamos calcular o valor da integral

$$ \int_0^\infty \frac{\sin(x)}{1 + x^3} dx. $$

```python
def f3(x):
    return np.sin(x) / (1 + x**3)
```

Aproxime as 30 primeiras integrais (novamente, usando `cheby16`)

$$ I_n = \int_{n\pi}^{(n+1)\pi} \frac{\sin(x)}{1 + x^3} dx $$

e acelere a sequência das somas parciais.

O código final pode ser pequeno, mas deve responder às perguntas:
- Como usar estas acelerações para aproximar o valor da integral?
- Como saber se o número de subdivisões para `cheby16` está apropriado?
- Como estimar o erro de sua resposta, apenas com os valores das integrais e das sequências aceleradas?

```python
ans = 0.6109127950469004

Is = [cheby16(f3, n*np.pi, (1+n)*np.pi, 9) for n in range(30)]
sin_poly = np.cumsum(Is)
cur = sin_poly
while True:
    print(cur - ans)
    if np.abs(cur[0] - cur[1]) < 1e-14:
        print(cur[0], cur[1])
        print(len(cur))
        break
    cur = aitken(cur)

aitken_plot(sin_poly, ans, maxiter=6)
plt.title("Erro de aproximação da integral")
plt.xlabel("Número de termos utilizados");
```
