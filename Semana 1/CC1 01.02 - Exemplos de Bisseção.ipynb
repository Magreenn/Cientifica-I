# Um polinômio de grau 5 $$x^5 + x - 1 = 0$$

$$ $$

$$ $$

$$ $$

$$ $$

$$ $$

$$ $$



$$ $$

$$ $$

```python
def poly5(x): return x**5 + x - 1
```

$$ $$

$$ $$

```python
xs = np.linspace(-2,2)
plt.plot(xs, poly5(xs))
plt.axhline(color="black")
plt.axvline(color="black");
```

```python
xs = np.linspace(-1,1.5)
plt.plot(xs, poly5(xs))
plt.axhline(color="black")
plt.axvline(color="black");
```

```python
r = bissection(poly5, 0, 1)
print(f"Raiz r: {r}, f(r) = {poly5(r)}")
```

```python
r, esquerda, direita = bissection(poly5, 0, 1, trace=True)
for e,d in zip(esquerda,direita):
    print(f"Esquerda: {e:.16f}, Raiz: {r}, Direita: {d:.16f}")
```

```python
plt.plot(esquerda, label="limite inferior")
plt.plot(direita,  label="limite superior")
plt.legend();
```

```python
ns = range(20, len(esquerda))
plt.plot(ns, esquerda[20:], label="limite inferior")
plt.plot(ns, direita[20:],  label="limite superior")
plt.gca().ticklabel_format(axis='y', useOffset=False)
plt.legend();
```

# Uma equação conhecida $$\cos(x) = x$$

$$ $$

$$ $$



```python
def f_cos(x): return x - np.cos(x)

r = bissection(f_cos, 0, 1)
print(f"Raiz r: {r}, f(r) = {f_cos(r)}, cos(r) = {np.cos(r)}")
```

```python
xs = np.linspace(-1, 1.5)
plt.plot(xs, xs, label="$x$")
plt.plot(xs, np.cos(xs), label="$\cos(x)$")
plt.axhline(color="black")
plt.axvline(color="black")
plt.scatter(r,r, s=100, color="C6")
plt.legend();
```

```python
xs = np.linspace(-1, 1.5)
plt.plot(xs, xs - np.cos(xs), label="$x - \cos(x)$")
plt.axhline(color="black")
plt.axvline(color="black")
plt.scatter(r,0, s=100, color="C6")
plt.legend();
```

```python
r = bissection(f_cos, 1, 3)
print(f"Raiz r: {r}, f(r) = {f_cos(r)}, cos(r) = {np.cos(r)}")
```

# Uma variante $$\sin(x) = \frac{x}{2}$$

$$ $$

$$ $$

```python
xs = np.linspace(-3, 4)
plt.plot(xs, xs/2, label="$x/2$")
plt.plot(xs, np.sin(xs), label="$\sin(x)$")
plt.axhline(color="black")
plt.axvline(color="black")
plt.legend();
```

```python
def f_sin(x): return np.sin(x) - x/2

r = bissection(f_sin, 1, 3)
print(f"Raiz r: {r}, f(r) = {f_sin(r)}, sin(r) = {np.sin(r)}, r/2 = {r/2}")
```

```python
r = bissection(f_sin, -3, -1)
print(f"Raiz r: {r}, f(r) = {f_sin(r)}, sin(r) = {np.sin(r)}, r/2 = {r/2}")
```

```python
r = bissection(f_sin, -1, -3)
print(f"Raiz r: {r}, f(r) = {f_sin(r)}, sin(r) = {np.sin(r)}, r/2 = {r/2}")
```

```python
r, esquerda, direita = bissection(f_sin, -3, -1, trace=True)

plt.plot(esquerda, label="limite inferior")
plt.plot(direita,  label="limite superior")
plt.legend();
```

```python
r = bissection(f_sin, -3, 4)
print(f"Raiz r: {r}, f(r) = {f_sin(r)}, sin(r) = {np.sin(r)}, r/2 = {r/2}")
```

```python
r = bissection(f_sin, -1, 2)
print(f"Raiz r: {r}, f(r) = {f_sin(r)}, sin(r) = {np.sin(r)}, r/2 = {r/2}")
```
