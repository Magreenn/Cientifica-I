```python
import numpy as np
import matplotlib.pyplot as plt
```

# Integrais difíceis

Vimos em aula que os métodos de "alta ordem" convergem mais rápido quando a função é várias vezes diferenciável.
Neste teste, vamos olhar a situação contrária:
vamos integrar funções que não são diferenciáveis no intervalo de integração,
e ver como os métodos se comportam!

```python
# Métodos de Cauchy, ponto médio e simpson.
# Nomes: cauchy, midpoint, simpson
# Forma: (f,a,b,n=100)

def rect(f,a,b,n=100):
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    return np.sum(f(l))*h

def midpoint(f,a,b,n=100):
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    m = l+h/2
    return np.sum(f(m))*h

def simpson(f,a,b,n=100):
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    m = l+h/2
    return (4*np.sum(f(m)) + 2*np.sum(f(l))+f(b)-f(a))*h/6
methods = [rect, midpoint, simpson]
```

## Parte 1: Integrando uma função que não é diferenciável

```python
f = np.abs
```

### 1: Primitiva, para calcular o erro de integração

Escreva a função `F(x)` que dá uma primitiva analítica de $f$.

```python
def F(x):
    if x>0:
        return (x**2)/2
    else:
        return -(x**2)/2
F= np.vectorize(F)
```

```python
# Para testar que é "mesmo" uma primitiva, com ótima precisão
def df(f,x,h=2**-17):
    """17*3 = 54"""
    return (f(x+h) - f(x-h))/(2*h)
```

```python
assert np.abs(df(F,1) - f(1)) < 1e-12
```

```python
assert np.abs(df(F,-2) - f(2)) < 2e-12
```

```python
np.random.seed(1)
xs = np.random.randn(10)
relerr = (df(F,xs) - f(xs))/f(xs)
assert np.all(np.abs(relerr) < 2e-11)
```

$F$ é diferenciável em zero, mas é uma conta mais difícil...

```python
df(F,0)
```

### 2: Erro em função do número de pontos

A que velocidade decai o erro de integração de $f$ no intervalo $[0,1]$?

1. Faça um gráfico com alguns valores de $n$, para observar a ordem dos três métodos
    - Dica: `f.__name__` dá o nome de uma função, para você usar na legenda
2. Depois, dê a ordem do método de Cauchy

```python
ans = F(1) - F(0)
ns = np.logspace(1,4,dtype=int)
for m in methods:
    y=[m(f,0,1,ni) for ni in ns]
    plt.loglog(ns,np.abs(y-ans),label=f"{m.__name__}")
    
plt.xlabel("Número de Passos")
plt.ylabel("Erro")
plt.title("Erro Absoluto dos Métodos")
plt.legend(title="Método")
plt.grid()
plt.show()
```

```python
# Forma da resposta: decay_speed = n
err_rect = [rect(f,0,1,ni) - ans for ni in ns]
ratio = [np.log(err_rect[i-1]/err_rect[i])/np.log(ns[i]/ns[i-1]) for i in range(len(ns))]
decay_speed = np.mean(ratio)
```

```python
for n in np.random.randint(100,10000, size=(6)):
    I1 = rect(f,0,1,n)
    err1 = np.abs(I1 - ans)
    I2 = rect(f,0,1,2*n)
    err2 = np.abs(I2 - ans)
    assert np.abs( err2/err1 - 2**-decay_speed ) < 2*n*1e-14
```

Agora, explique o que aconteceu nos métodos do ponto médio e de Simpson.

Sabe-se que o metodo de $\frac{1}{3}$ Simpson Repetida tem erro da forma: $$\max{|f^{iv}(x)|}\frac{(b-a)h^{4}}{180}$$ e o erro do método do Ponto Médio tem a forma: $$\max{|f^{''}(x)|}\frac{(b-a)h^{2}}{24}$$. 

E por estarmos aplicando esses métodos numa função polinomial de ordem 1 as formulas dos erros para os métodos dão $0$ já que a função tem suas derivadas das respectivas ordens iguais a $0$ tambem. 

Há de se comentar o problema de não existir a derivada no ponto $0$, que faz parte do intervalo, porem este problema não influencia no método já que o intervalo só é definido à direita desse ponto, podendo ser entendido como um "limite lateral direito". 

### 3. Mudando o intervalo de integração

Agora, faça o gráfico do erro para o intervalo $[-2,1]$.

```python
# Faça aqui uma função genérica para "fazer gráficos de erro" para os 3 métodos conjuntamente
# Não esqueça das legendas!
def graph_err(f,a,b,ans):
    ns = np.logspace(1,4,dtype=int)
    for m in methods:
        y=[m(f,a,b,ni) for ni in ns]
        plt.loglog(ns,np.abs(y-ans),label=f"{m.__name__}")
    plt.xlabel("Número de Passos")
    plt.ylabel("Erro")
    plt.title(f"Erro Absoluto dos Métodos no Intervalo $[{a:.2f},{b:.2f}]$")
    plt.legend(title="Método",loc="upper left", bbox_to_anchor=(1.0,1.0))
    plt.grid();
```

```python
a,b = -2,1
ans = F(b) - F(a)
graph_err(f,a,b,ans)
plt.show()
```

Quais são as velocidades de convergência dos métodos?

Você percebe algum comportamento especial?  Como explicar isso?

```python
a,b = -2,1
ans = F(b) - F(a)
ns = np.logspace(1,4,dtype=int)
for m in [rect, midpoint]:
    err = [m(f,a,b,ni) - ans for ni in ns]
    decay_speed = np.log(np.abs(err[0]/err[-1]))/np.log(np.abs(ns[-1]/ns[0]))
    print(f"Velocidade de Convergência do Método {m.__name__} é {decay_speed}")
```

Podemos ver que o erro para o método do Ponto Médio mudou de comportamento e os outros dois se mantiveram. Um motívo para essa mudança de comportamento é que, ao mudar o intervalo de integração, o ponto onde a derivada não existe agora está no meio do interválo, então podemos ver que a fórmula teórica do erro passa a dar a forma do gráfico, porem sem a derivada segunda: $$E_{PM} = \frac{(b-a)h^{2}}{24}$$

Para o método de Simpson não ocorre o mesmo fenômeno pois para este método é utilizado uma aproximação do intervalo por uma parábola usando 3 pontos e o 0 não é utilizado para isso.

Agora, repita o estudo para o intervalo $[-1,1]$.

Aqui, será melhor ter um gráfico separado para cada um dos três métodos.

```python
ans = F(1) - F(-1)
ns = np.logspace(1,4,dtype=int)
_, axs = plt.subplots(ncols=3, figsize=(15,4), sharey=True)
for m,ax in zip(methods,axs):
    y = [m(f,-1,1,ni) for ni in ns]
    ax.loglog(ns,np.abs(y-ans))
    ax.set_xlabel("Número de Passos")
    ax.set_ylabel("Erro")
    ax.set_title(f"Erro Absoluto do Método {m.__name__}")
    
plt.show()
```

O que aconteceu agora?
Porquê?

Podemos ver claramente que todos os métodos não tiveram o comportamento teórico esperado das suas fórmulas de erros máximos, já que todos tem erros de ordem $O(h)$. 
Este comportamento pode ser explicado pois, a função analizada é uma reta e esta sendo integrada em um intervalo simétrico, sendo assim não há ganho de precisão por calcular pontos intermediários no método de simpson, os três métodos estão fazendo exatamente a mesma operação com mais ou menos pontos.

### 4. Integrando em intervalos "aleatórios"

Aqui, usar a `graph_err` vai ser uma boa ideia ;-)

```python
_, axs = plt.subplots(ncols=3, figsize=(15,4))
for ax in axs:
    # Três intervalos aleatórios!
    a,b = -np.random.rand(), np.random.rand()
    ax.set_title('$|x|$ in $[{:.2},{:.2}]$'.format(a,b))
    plt.sca(ax)
    graph_err(f,a,b,(F(b)-F(a)))
    
plt.show()
```

Estes resultados são mais fáceis ou mais difíceis de interpretar?
Porquê?

Esses resultados são mais difícieis de interpretar pois a escolha aletória do intervalo torna dificil observar a ordem de convergencia de cada método, pois há muito barulho nos gráficos devido a escolha descriterioza dos intervalos de integração o que afeta diretamente no passo dos métodos e nos erros dos mesmos.
