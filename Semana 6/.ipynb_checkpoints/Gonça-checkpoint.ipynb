# Integrais difíceis

Vimos em aula que os métodos de "alta ordem" convergem mais rápido quando a função é várias vezes diferenciável.
Neste teste, vamos olhar a situação contrária:
vamos integrar funções que não são diferenciáveis no intervalo de integração,
e ver como os métodos se comportam!

```python
# Métodos de Cauchy, ponto médio e simpson.
# Nomes: cauchy, midpoint, simpson
# Forma: (f,a,b,n=100)

def rect(f,a,b,n=100,pontos=False):
    
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    if pontos==True:
        return l,np.sum(f(l))*h
    else:
        return np.sum(f(l))*h

def midpoint(f,a,b,n=100,pontos=False):
    
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    mids = l + h/2
    if pontos==True:
        return mids,np.sum(f(mids))*h
    else:
        return np.sum(f(mids))*h


def simpson(f,a,b,n=100,pontos=False):
    
    l,h = np.linspace(a,b,num=n,endpoint=False,retstep=True)
    mids = l + h/2
    if pontos==True:
        return l,mids,(4*np.sum(f(mids))+2*np.sum(f(l))+(f(b)-f(a)))*(h/6)
    else:
        return (4*np.sum(f(mids))+2*np.sum(f(l))+(f(b)-f(a)))*(h/6)

methods = [rect, midpoint, simpson]
```

## Parte 1: Integrando uma função que não é diferenciável

```python
f = np.abs
```

### 1: Primitiva, para calcular o erro de integração

Escreva a função `F(x)` que dá uma primitiva analítica de $f$.

```python
def F(x):
    if x>0:
        return (x**2)/2
    else:
        return -(x**2)/2
    
F = np.vectorize(F);
```

```python
# Para testar que é "mesmo" uma primitiva, com ótima precisão
def df(f,x,h=2**-17):
    """17*3 = 54"""
    return (f(x+h) - f(x-h))/(2*h)
```

```python
assert np.abs(df(F,1) - f(1)) < 1e-12
```

```python
assert np.abs(df(F,-2) - f(2)) < 2e-12
```

```python
np.random.seed(1)
xs = np.random.randn(10)
relerr = (df(F,xs) - f(xs))/f(xs)
assert np.all(np.abs(relerr) < 2e-11)
```

$F$ é diferenciável em zero, mas é uma conta mais difícil...

```python
df(F,0)
```

### 2: Erro em função do número de pontos

A que velocidade decai o erro de integração de $f$ no intervalo $[0,1]$?

1. Faça um gráfico com alguns valores de $n$, para observar a ordem dos três métodos
    - Dica: `f.__name__` dá o nome de uma função, para você usar na legenda
2. Depois, dê a ordem do método de Cauchy

```python
ans = F(1) - F(0)
ns = np.logspace(1,4,dtype=int)
for m in methods:
    t= [m(f,0,1,n) for n in ns]
    plt.loglog(ns,abs(t-ans),label=f"{m.__name__}")
    
plt.xlabel("Número de Passos")
plt.ylabel("Erro")
plt.title("Gráficos dos Erros")
    
plt.legend()
plt.grid()
plt.show() 
```

```python
# Forma da resposta: decay_speed = n
# YOUR CODE HERE

ans = F(1) - F(0)
ns = np.logspace(1,4,dtype=int)
Err_rect = [(rect(f,0,1,n) - ans) for n in ns]
Decaimento_Passo = []
for i in range(len(ns)-1):
    Decaimento_Passo.append(np.log(Err_rect[i]/Err_rect[i+1])/np.log(ns[i+1]/ns[i]))

decay_speed = np.mean(Decaimento_Passo)
```

```python
for n in np.random.randint(100,10000, size=(6)):
    I1 = rect(f,0,1,n)
    err1 = np.abs(I1 - ans)
    I2 = rect(f,0,1,2*n)
    err2 = np.abs(I2 - ans)
    assert np.abs( err2/err1 - 2**-decay_speed ) < 2*n*1e-14
```

Agora, explique o que aconteceu nos métodos do ponto médio e de Simpson.

YOUR ANSWER HERE

Como se sabe, o erro para a regra de Simpson Repetida depende de $h^4$ assim como da quarta derivada da função.

Da mesma forma, o erro para a regra do Ponto Médio depende de $h^2$ assim como da segunda derivada da função.

Entre o intervalo de 0 à 1 o problema da ausência de derivada no ponto 0 não influencia nos métodos, já que o intervalo só está definido à direita do ponto. Sendo assim, no ponto zero é "visto" apenas o limite lateral positivo, o qual é 1. 

Dessa forma, considerando todo o intervalo à direita do ponto 0, por ser uma função polinominal de ordem 1, esta terá segunda e quarta derivadas iguais à zero. Por isso, o erro para o ponto médio e para Simpson será dependente apenas do erro de truncamento.

### 3. Mudando o intervalo de integração

Agora, faça o gráfico do erro para o intervalo $[-2,1]$.

```python
# Faça aqui uma função genérica para "fazer gráficos de erro" para os 3 métodos conjuntamente
# Não esqueça das legendas!
def graph_err(f,a,b,ans):
    ns = np.logspace(1,4,dtype=int)
    for m in methods:
        t= [m(f,a,b,n) for n in ns]
        plt.loglog(ns,abs(t-ans),label=f"{m.__name__}")
    plt.xlabel("Número de Passos")
    plt.ylabel("Erro")
    plt.title(f"Gráficos dos Erros no intervalo {a: .2f} e {b: .2f}")
    plt.legend()
```

```python
a,b = -2,1
ans = F(b) - F(a)
graph_err(f,a,b,ans)
plt.show()
```

Quais são as velocidades de convergência dos métodos?

Você percebe algum comportamento especial?  Como explicar isso?

YOUR ANSWER HERE  

A convergência do método do retângulo continua na ordem de $h^1$ e o método de simpson continua apresentando erros apenas de truncamento para todo o intervalo. 

Entretanto, neste intervalo nota-se que o método do ponto médio passa a ter erros de aproximação na ordem de $h^2$, visto que dentro do intervalo a função possui um ponto onde a derivada não existe. Nota-se que o erro é limitado pela sua fórmula desconsiderando a parcela da derivada (ou seja, com a derivada igual à 1): 

$$Erro_{Ponto Médio} = \frac{(b-a) \times h^2}{24}$$

Para a função de simpson isto não ocorre pois a função se baseia na "interpolação" da função com três pontos de aproximação, e no intervalo de -2 à 1 o ponto 0 não é utilizado.

```python
def erro_mid(a,b,n):
    return (1/24)*(((b-a)/n)**2)*(b-a)

a,b = -2,1
ans = F(b) - F(a)

t= [midpoint(f,a,b,n) for n in ns]
err_real = [erro_mid(a,b,n) for n in ns]

plt.loglog(ns,abs(t-ans),label="Erro Real Algoritmo");
plt.loglog(ns,err_real,label="Erro Teórico");
plt.legend();
plt.xlabel("Número de Passos");
plt.ylabel("Erro");
```

```python
ans = F(1) - F(-2)
ns = np.logspace(1,4,dtype=int)
for m in [rect,midpoint]:
    Err_rect = [(m(f,-2,1,n) - ans) for n in ns]

    decay_speed = np.log(Err_rect[0]/Err_rect[-1])/np.log(ns[-1]/ns[0])
     
    print(f"A velocidade de convergência aproximada do método {m.__name__} é da ordem de h**{decay_speed}");
```

```python
pts_ex,pts_mid,_ = simpson(f,-2,1,n=100,pontos=True)
print(f"Os pontos dos extremos no método de Simpson para o intervalo -2 à 1 são \n \n {pts_ex} \n")
print(f"Os pontos médios no método de Simpson para o intervalo -2 à 1 são \n \n {pts_mid}")
```

Agora, repita o estudo para o intervalo $[-1,1]$.

Aqui, será melhor ter um gráfico separado para cada um dos três métodos.

```python
ans = F(1) - F(-1)
ns = np.logspace(1,4,dtype=int)
_, axs = plt.subplots(ncols=3, figsize=(15,4), sharey=True)
for m,ax in zip(methods,axs):
        t= [m(f,-1,1,n) for n in ns]
        ax.loglog(ns,abs(t-ans))
        ax.set_xlabel("Número de Passos")
        ax.set_ylabel("Erro")
        ax.set_title(f"Erro do gráfico {m.__name__}")
        
plt.show()
```

O que aconteceu agora?
Porquê?

YOUR ANSWER HERE

O interessante neste caso é que nenhum dos métodos seguem o limite máximo teórico, estando todos os erros em ordem de h. Com um intervalo simétrico em relação ao eixo y todos os intervalos possuem a mesma ordem de erro. 

No caso do retângulo, isto acontece pois como o intervalo é simétrico e a função também é simétrica em relação ao eixo y, à esquerda do zero o método superestima a função, à direita o método subestima a integral. Dessa forma, os erros acabam se cancelando, fazendo com que o erro fique da ordem de $h^2$. 

Esta forma de "compensação do erro" devido as características deste caso também ocorre para o método de simpson. Como a função módulo é apenas uma reta espelhada ao redor do eixo y, o método de simpson acaba na prática fazendo o mesmo que o ponto médio. Como exemplo didático, vamos supor que estamos integrando -1 à 1 com apenas 2 intervalos. O primeiro intervalo será dado por:

$$I(1) = \frac{(f(-1) + 4 \times f(-0.5) + f(0))*h}{6}$$

$$I(1) = \frac{(1 + 4 \times 0.5 + 0)*h}{6}$$

$$I(1) = \frac{3*h}{6}$$

O que na prática é:

$$I(1) = \frac{f(\frac{x_i + x_{i+1}}{2})*h}{2}$$

E isto na prática é o mesmo que o ponto médio. Como o intervalo de integração é simétrico (-1 à 1) e a função é simétrica em relação ao eixo y, o mesmo ocorrerá para o segundo intervalo - de 0 à 1.

```python
def erro_mid(a,b,n):
    return (1/24)*(((b-a)/n)**2)*(b-a)

def erro_rect(a,b,n):
    return (1/2)*(((b-a)/n))*(b-a)

def erro_simp(a,b,n):
    return (1/2880)*(((b-a)/n)**4)*(b-a)

fig, axs = plt.subplots(ncols=3, figsize=(15,4))

a,b = -1,1
ans = F(b) - F(a)
Erros = [erro_rect,erro_mid,erro_simp]

for m,ax,err_m in zip(methods,axs,Erros):

    t= [m(f,a,b,n) for n in ns]
    err_real = [err_m(a,b,n) for n in ns]

    ax.loglog(ns,abs(t-ans),label="Erro Real Algoritmo");
    ax.loglog(ns,err_real,label="Erro Teórico");
    ax.set_xlabel("Número de Passos");
    ax.set_ylabel("Erro");
    ax.set_title(f"Erro do gráfico do {m.__name__}")#
```

### 4. Integrando em intervalos "aleatórios"

Aqui, usar a `graph_err` vai ser uma boa ideia ;-)

```python
_, axs = plt.subplots(ncols=3, figsize=(15,4))
for ax in axs:
    # Três intervalos aleatórios!
    a,b = -np.random.rand(), np.random.rand()
    ax.set_title('$|x|$ in $[{:.2},{:.2}]$'.format(a,b))
    plt.sca(ax)
    graph_err(f,a,b,(F(b)-(F(a))))
    
plt.show()
```

Estes resultados são mais fáceis ou mais difíceis de interpretar?

Entretanto, esta aferição neste caso se torna mais difícil pela alta volatilidade observada.
